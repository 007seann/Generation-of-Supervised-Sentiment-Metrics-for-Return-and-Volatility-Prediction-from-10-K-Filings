```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
ten_k_data_dir = "/Users/apple/PROJECT/Code_4_SECfilings/fillings/company_df"
# get all the files in the directory
ten_k_files = os.listdir(ten_k_data_dir)

ten_k_reports = []

for file in tqdm(ten_k_files):
    if file.endswith('.csv'):
        ten_k_reports.append(pd.read_csv(os.path.join(ten_k_data_dir, file)))

ten_k_reports = pd.concat(ten_k_reports)
# reset index
ten_k_reports.reset_index(drop=True, inplace=True)
# rename "Name" column to "symbol", "Date" to "date", "Body" to "10K_text"
ten_k_reports.rename(columns={"Name": "symbol", "Date": "date", "Body": "10k_text"}, inplace=True)
```

```python
ten_k_reports
```

```python
ten_k_reports.to_parquet("NASDAQ_100_10-K.parquet")
```
