```python
import os
root_fold_data = '/Users/apple/PROJECT/Code_4_10k/data'
data_folder_file_counts = {}

for folder_name in os.listdir(root_fold_data):
    folder_path = os.path.join(root_fold_data, folder_name)
    if os.path.isdir(folder_path):
        files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
        data_folder_file_counts[folder_name] = len(files)
print(data_folder_file_counts)
print(len(data_folder_file_counts))
data_num_collected = 0    
for key, value in data_folder_file_counts.items():
    data_num_collected += value

print(data_num_collected)    
```
