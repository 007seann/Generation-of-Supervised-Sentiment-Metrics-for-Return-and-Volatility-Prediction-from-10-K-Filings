```python
import pandas as pd
```

```python
df = pd.read_csv('Loughran-McDonald_10X_Summaries_1993-2021.csv')
```

```python
df
```

```python
# This code extracts only top 10 companies' 10-K meta data listed in QQQ ETF from 2007 to 2021

company_name = ["Apple Inc.", "APPLE INC", "APPLE COMPUTER INC", "MICROSOFT CORP", "AMAZON COM INC", "AMAZON SCIENTIFIC INC", "AMAZON BIOTECH INC", "NVIDIA CORP/CA", "NVIDIA CORP", "Facebook Inc", "Broadcom Inc.", "Broadcom Ltd", "Broadcom Cayman L.P", "Tesla  Inc.", "TESLA MOTORS INC", "Alphabet Holding Company  Inc.", "Alphabet Inc.", "COSTCO WHOLESALE CORP", "COSTCO WHOLESALE CORP /NEW"]

df = df[df['CoName'].isin([name for name in company_name])]
# Assuming df is your DataFrame
# Convert the FILING_DATE column to datetime if it's not already
df['FILING_DATE'] = pd.to_datetime(df['FILING_DATE'], format='%Y%m%d', errors='coerce')

# Define the date range
start_date = pd.Timestamp('2007-01-01')
end_date = pd.Timestamp('2021-12-31')

# Apply the date range filter
df = df[(df['FILING_DATE'] >= start_date) & (df['FILING_DATE'] <= end_date)]

# Filter by FORM_TYPE
filtered_df = df[df['FORM_TYPE'] == '10-K']
filtered_df
```

```python
files_num = filtered_df['ACC_NUM'].tolist()
files_num
```

```python
# Convert the 'CoName' column to a list
company_names = set(df['CoName'])
# Open a text file for writing
with open('company_names.txt', 'w') as file:
    for name in company_names:
        file.write(name + '\n')

print("Company names have been written to company_names.txt")
```
